## Elbow Technique

```python
## Number of Potential Clusters in Dataset
number_cl=range(1,10)
```

```python
KMeans1= [KMeans(n_clusters=i) for i in number_cl] # defines range for KMeans clusters
```

```python
## Create Score that Shows Variation for Each Number of Clusters
cluster_score = [KMeans1[i].fit(df).score(df) for i in range(len(KMeans1))]
```

```python
cluster_score
```

```python
## Create Elbow Rule Plot
pl.plot(number_cl, cluster_score) ## Plot Clusters and Score Variables
pl.xlabel('Number of Clusters')  ## X Axis
pl.ylabel('Cluster Score') ## Y Axis
pl.title('Elbow Curve')  ## Give a Title
pl.show()  ## Show the Plot
```

The optimal number of clusters is 3 because the marginal difference in the cluster score starts to drop after that.
However, an argument can be made for going up to 4 based on the graph. There is still an decent difference in the cluster score from 3 to 4.

## K Means Clustering

#### Create Object of K Means


```python
KMeans2=KMeans(n_clusters=3)
```

```python
## Fit 3 K Means Clusters to Data
KMeans2.fit(df)
```

#### Create Clusters Column

```python
df['clusters']=KMeans2.fit_predict(df)

## Show Headers to Confirm
df.head()
```

#### Check Value Counts

```python
df['clusters'].value_counts()
```

#### Get Mean and Median of Each Cluster

```python
df.groupby('cluster').agg({
'Expected Years of Schooling': ['mean','median'],
                           'Gender Dev Index (Gender Inequality)': ['mean','median'],
                           'Gross National Income Per Capita': ['mean','median'],
                          'Human Development Index': ['mean','median'],
                          'Labour force participation rate, female % ages 15 and older)': ['mean', 'median'],
                          'Life Expectancy': ['mean','median'],
                          'Maternal Mortality Ratio (deaths per 100,000 live births)': ['mean', 'median']
})                           
```

I will be discussing means and medians for each section due to the numerous variables here. The median value for expected years of schooling is higher than the mean for each of the clusters. This means that there are some very small values for life expectancy that is dragging down the mean. 
For Gender Development Index, the medians and medians are about even, with the exception for the tan cluster with the mean being lower than the median, meaning some much smaller GDIs weighing things down. 
For GNI Per Person, Dark Purple has roughly the same mean and median. For purple and tan, the mean is substantially greater than the median, meaning that for lower and middle income countries, there are some really large values that are tipping the scale in favor of the mean.
For HDI, dark purple and purple are roughly the same with tan having a lower mean.
With Labor Force Participation Rate for Females Aged 15 and Over, mean and median are about the same, meaning large values don't play a role here.
For Life Expectancy, for dark purple and purple, mean and median are about the same with large values not playing a role. When looking at Tan, the mean is slighly less, meaning a few large values are tipping the scale in favor of the median being larger. 
Maternal Mortality (Deaths Per 100k Live Births) has the largest difference between mean and median, each cluster having a mean more than twice as large as the median. This means there are a significant number of extremely large values that are tipping the scales. What is interesting here is that these are for all of the clusters and not just the cluster for the lowest developed countries (tan).
